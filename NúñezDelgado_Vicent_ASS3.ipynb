En esta práctica implementaremos un algoritmo de ramificación y poda para solucionar el problema de la mochila:

(from Wikipedia): En algoritmia, el problema de la mochila, comúnmente abreviado por KP (del inglés Knapsack problem) es un problema de optimización combinatoria, es decir, que busca la mejor solución entre un conjunto finito de posibles soluciones a un problema. Modela una situación análoga al llenar una mochila, incapaz de soportar más de un peso determinado, con todo o parte de un conjunto de objetos, cada uno con un peso y valor específicos. Los objetos colocados en la mochila deben maximizar el valor total sin exceder el peso máximo.

Supongamos que tenemos n distintos tipos de ítems, que van del 1 al n. De cada tipo de ítem se tienen q ítems disponibles (que puede ser diferente para cada ítem).

Cada tipo de ítem tiene un beneficio asociado v y un peso (o volumen) w. Se asume que el beneficio y el peso no son negativos. 

Por otro lado se tiene una mochila, donde se pueden introducir los ítems, que soporta un peso máximo (o volumen máximo) W.

El problema consiste en meter en la mochila ítems de tal forma que se maximice el valor de los ítems que contiene y siempre que no se supere el peso (o volumen) máximo que puede soportar la misma. La solución al problema vendrá dado por la secuencia de variables x1, x2, ..., xn donde el valor de xi indica cuantas copias se meterán en la mochila del tipo de ítem i.

<h3>INPUT</h3>
<ul>
<li>elements: Diccionario que contiene una clave con el nombre del item, y como valor una tupla de tres elementos que contienen (Peso, Valor, Cantidad).  
</li>
<li>total_weight: peso máximo que soporta la mochila a ser rellenada por ítems con el objetivo de maximizar el valor que contiene.</li>
</ul>
<br>
<h3>OUTPUT</h3>
Listado de elementos que se introducen en la mochila (en el orden en el que se introducen, índice del ítem, la longitud del vector es la cantidad de elementos que se introducen en la mochila) y el valor total de la mochila.

```python
def branch_bound(elements, total_weight):
    
    items=[]
    valor=0
    
    lista_de_items=[]
    for element in elements:
        for i in range(0,elements[element][2]):
            lista_de_items.append((element,elements[element][0],elements[element][1]))
            
   
    n=(len(lista_de_items))/(len(lista_de_items[0]))
    
    
    print(knapsack(total_weight,lista_de_items,n))
    
def bound(u, n,W, lista_de_items):
    
    #Si el peso sobrepasa la capacidad devulve
    # 0 como cota
    if u[3] >= W:
      
        return 0;

    # inicializar cota beneficio por la cota actual
    profit_bound = u[1];

    #start including items from index 1 more to current
    #item index
    j = u[0] + 1;
    totweight = u[3];

    # checking index condition and knapsack capacity
    #condition
    while j < n and  totweight + lista_de_items[j][1] <= W :
    
        totweight    += lista_de_items[j][1];
        profit_bound += lista_de_items[j][2];
        j=j+1
    

     #If k is not n, include last item partially for
     #upper bound on profit
    if  j < n :
        profit_bound += (W - totweight) * lista_de_items[j][2] / lista_de_items[j][1];

    return profit_bound;

    
    
def knapsack(W,lista_de_items,n):
    
   
    
    # make a queue for traversing the node
    Q=[]
    
    #u=(nivel,provecho,upperbound,peso)
    u=[]
    v=[]
    listaDeItems=[]
    
    u.append(-1)
    u.append(0)
    u.append(0)
    u.append(0)
    u.append('')
    Q.append(u)
    # Uno a uno extrae un elemento del árbol de decisiones
    #calcula el beneficio de todos los hijos del elemento extraido
    #y seguir guardando maxProfit
    
    #v=(nivel,provecho,upperbound,peso)
    v.append(0)
    v.append(0)
    v.append(0)
    v.append(0)
    v.append('')
    
    maxProfit = 0;
    
    while (len(Q)!=0  ):
   
    
        #Dequeue un  nodo
        
        u = Q[0];
        del Q[0];
       
        #Si es nodo inicial, asigna el nivel 0
        if  u[0] == -1:
                v[0]=0
               
        #Si no hay nada en el siguiente nivel continua
        if  u[0] == n-1:
             continue;

         #Si no es el último nodo, luego incrementa el nivel,
         #y calcular el beneficio de los nodos hijos.
        v[0] = u[0] + 1;
        
        #Tomando el elemento del nivel actual, agrega 
        #el peso del nivel actual y el valor para el nodo peso y el valor de u
        
        v[4] = lista_de_items[v[0]][0];
      
        v[3] = u[3] + lista_de_items[v[0]][1];
        v[1] = u[1]+ lista_de_items[v[0]][2];
        
       
        
        
        
        # Si el peso acumulado es menor que W y
        #profit es mayor que el beneficio anterior,
        # actualización maxprofit
        if (v[3] <= W and v[1] > maxProfit):
            maxProfit = v[1];
            listaDeItems.append(v[4])

        # calcula el `bound`
        
        v[2] = bound(v, n, W, lista_de_items);

        # Si 'bound' es mayor que la cota,
        # entonces se añade a la cola para comprobarlo mas tarde
        
        if (v[2] > maxProfit):
            print("Q:")
            print(Q)
            Q.append(v)

        # Hace lo mismo, pero sin tomar
        # el objeto en mochila
        v[3] = u[3];
        v[1]= u[1];
        v[2] = bound(v, n, W, lista_de_items);
        
        if v[2] > maxProfit:
           
            Q.append(v);
            
    
    print("Weight total :")
    print(u[3])
    
    return maxProfit,listaDeItems;
```

```python
elements = {
    'lamp': (10, 100, 3), # Weight, Value, Amount
    'rock': (100, 1, 40), # Weight, Value, Amount
    'match': (1, 80, 20), # Weight, Value, Amount
    'meat': (5, 100, 5), # Weight, Value, Amount
    'shirt': (3, 20, 2), # Weight, Value, Amount
    'tea': (2, 15, 60), # Weight, Value, Amount
}
branch_bound(elements,1000)
```
